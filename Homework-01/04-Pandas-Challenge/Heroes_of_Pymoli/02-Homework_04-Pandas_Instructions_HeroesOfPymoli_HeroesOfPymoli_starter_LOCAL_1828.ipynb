```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data.head()
```

## Player Count

* Display the total number of players


```python
unique_players = len(set(purchase_data.iloc[0:,1]))

table = {'Total Players': unique_players}
print_table = pd.DataFrame(table, columns=['Total Players'], index=[0])
print_table
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
group = purchase_data.groupby('Item Name').Price.sum()
total_items = len(group)
total_revenue = sum(group)
average_price = purchase_data.Price.mean()
total_purchases = purchase_data.Price.count()

analysis_table = {
    'Number of Unique Items': total_items,
    'Average Price': average_price,
    'Number of Purchases': total_purchases,
    'Total Revenue': total_revenue
}

display_table = pd.DataFrame(analysis_table, columns=['Number of Unique Items',
    'Average Price',
    'Number of Purchases',
    'Total Revenue'],index=[0])
display_table
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_reveal = purchase_data.groupby('Gender')

unique_gender = gender_reveal['SN'].nunique()

percentage = ((unique_gender/unique_gender.sum()) * 100)

df = {
    'Total Count': unique_gender,
    'Percentage of Players': percentage
}

print_df = pd.DataFrame(df)
print_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
genders = purchase_data.groupby('Gender')
count = len(purchase_data['SN'].value_counts())

# set variables to calculate
purchase_gender = genders.count()['Price']
average_purchase = (genders['Price'].mean()).map('${:.2f}'.format)
total_purchase = purchase_data.groupby(['Gender']).sum()['Price']
avg_total_person = total_purchase/print_df['Total Count']

# make df
purchase_analysis_df = pd.DataFrame({
    'Purchase Count': purchase_gender,
    'Average Purchase Price': average_purchase,
    'Total Purchase Value': total_purchase,
    'Avg Total Purcahse per Person': avg_total_person
})

# round to 2 decimals
purchase_analysis_df = purchase_analysis_df.round(2)

# print df
purchase_analysis_df.sort_index()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
users = purchase_data.loc[:, ['SN','Age','Gender']]
users = users.drop_duplicates()

# Create age bins
age_bin = [0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 999]
bin_labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

# Separate players in bins
users['Age Ranges'] = pd.cut(users['Age'], age_bin, labels = bin_labels)

# Calculate
demo_totals = users['Age Ranges'].value_counts()
demo_percents = (demo_totals / count) * 100

# Create df
demo_df = pd.DataFrame({'Total Count': demo_totals,
                       'Percentage of Players': demo_percents})

# Round to 2 decimals
demo_df = demo_df.round(2)

# Print df
demo_df.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#create df basesd on age and price
users_price = purchase_data.loc[:, ['SN','Age','Price']]


# create bins
age_bin = [0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 999]
bin_labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

# separate players in bins
users_price['Age Ranges'] = pd.cut(users_price['Age'], age_bin, labels = bin_labels)

# calculate
purchase_count = users_price['Age Ranges'].value_counts()
average_purchase_price = users_price.groupby('Age Ranges').Price.mean()
total_purchase_value = users_price.groupby('Age Ranges').Price.sum()
total_purchase_person = users_price.groupby('Age Ranges').Price.sum()/demo_totals

# make df
purchase_df = pd.DataFrame({
    'Purchase Count': purchase_count,
    'Average Purchase Price': average_purchase_price,
    'Total Purchase Value': total_purchase_value,
    'Avg Total Purchase per Person': total_purchase_person
})

# round df to 2 decimals
purchase_df = purchase_df.round(2)

# print df
purchase_df.sort_index()
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# create main df
user_spending_df = purchase_data.loc[:, ['SN','Price']]
unique_user_spending = user_spending_df.groupby('SN')

# run calculations
purchase_count = purchase_data['SN'].value_counts()
avg_purchase_price = unique_user_spending['Price'].mean().round(2)
total_purch = user_spending_df.groupby('SN').Price.sum()

# make df
spender_df = pd.DataFrame({
    'Purchase Count': purchase_count,
    'Avg Purchase Price': avg_purchase_price,
    'Total Purchase': total_purch
})

# round df to 2 decimals
spender_df.round(2)

# print df
spender_df.sort_index(ascending=False)
```
